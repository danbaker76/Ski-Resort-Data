```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Importing Ski Resort Dataset
df = pd.read_csv('Ski_resort.csv')
```

```python
## Converting first date field to datetime
df['Date_of_purchase'] = pd.to_datetime(df['Date_of_purchase'],format='%Y-%m-%d')
```

```python
## Converting Second Date Field (initial day) to datetime
df['Initial_day'] = pd.to_datetime(df['Initial_day'],format='%Y-%m-%d')
```

```python
plt.figure(figsize = [10,6])
sns.countplot(data=df,x='Original_Resort',hue='Bought_in')
```

```python
## Breaking out the month & year to create month/year column
df['Purchase_month'] = pd.DatetimeIndex(df['Date_of_purchase']).month
df['Purchase_year'] = pd.DatetimeIndex(df['Date_of_purchase']).year
df['purchase_month_year'] = pd.to_datetime(df['Date_of_purchase']).dt.to_period('M')
```

```python
plt.figure(figsize = [20,6])
sns.countplot(data=df,x=df['purchase_month_year'].sort_values())
```
